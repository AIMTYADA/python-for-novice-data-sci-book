## **Summarizing and Computing Descriptive Statistics**

```python
df = pd.DataFrame([[1.4, np.nan], [7.1, -4.5],
                   [np.nan, np.nan], [0.75, -1.3]],
                  columns=['one', 'two'])
df
```

```python
# DataFrame sum(method) returs a Series containing column sums
df.sum()
```

```python
# summation across column instead
df.sum(axis='columns')
```

```python
# look at index with NaN, normally NaN values are skipped
df.mean(axis='columns')
```

```python
# Disable skip NaN with skipna=False
df.mean(axis='columns', skipna=False)
```

```python
# Some methods, like idxmin and idxmax, return indirect statistics like the index value where the minimum or maximum values are attained:
df.idxmax()
```

```python
# producing multiple summary statistics in one shot
df.describe()
```

## **Data Cleaning and Preparation**



### **Handling Missing Data**

*   Not-a-Number (NaN) 





```python
string_data = pd.Series(['aardvark', 'artichoke', np.nan, 'avocado'])
string_data
```

```python
string_data.isnull()
```

> NA handling methods
>* dropna() Filter axis labels based on whether values for each label have missing data, with varying thresholds for how
much missing data to tolerate.
>* fillna() Fill in missing data with some value or using an interpolation method such as 'ffill' or 'bfill'.
>* isnull() Return boolean values indicating which values are missing/NA.
>* notnull() Negation of isnull.

**Filter Out Missing Data**

```python
from numpy import nan as NA

data = pd.Series([1, NA, 3.5, NA, 7])
data
```

```python
data.dropna()
```

```python
# same result
data[data.notnull()]
```

```python
# DataFrame object
data = pd.DataFrame([[1., 6.5, 3.], [1., NA, NA],
                     [NA, NA, NA], [NA, 6.5, 3.]])
data
```

```python
# Drop and row containing NaN
cleaned = data.dropna()
cleaned
```

```python
# Only drop the rows with all NaN
data.dropna(how='all')
```

```python
# Add new Column
data[4] = NA
data
```

```python
# To drop column, pass parameter axis=1
data.dropna(axis=1, how='all')
```

### **Filling in Missing Data**

```python
# Create New DataFrame with some NaN
df = pd.DataFrame(np.random.randn(7, 3))
df.iloc[:4, 1] = NA
df.iloc[:2, 2] = NA
df
```

```python
df.fillna(0)
```

```python
# Fill in different value for each column
# Column 1: fillin 0.5
# Column 2: fillin 0
df.fillna({1: 0.5, 2: 0})
```

```python
# Create new df
df = pd.DataFrame(np.random.randn(6, 3))
df.iloc[2:, 1] = NA
df.iloc[4:, 2] = NA
df
```

```python
df.mean()
```

```python
# fill in using dataFrame's mean
df.fillna(df.mean())
```

```python
#fill in using interpolation methods
df.fillna(method='ffill')
```

## **Data Transformation**

### **Removing Duplicates**

```python
data = pd.DataFrame({'k1': ['one', 'two'] * 3 + ['two'],
                     'k2': [1, 1, 2, 3, 3, 4, 4]})
data
```

The DataFrame method duplicated returns a boolean Series indicating whether each row is a duplicate (has been observed in a previous row) or not:

```python
data.duplicated()
```

```python
data.drop_duplicates()
```

```python
data.drop_duplicates(keep='last')
```

```python
# Add a column
data['v1'] = range(7)
data
```

```python
# filter out duplicates in 'k1' column
data.drop_duplicates(['k1'])
```

### Transforming Data Using a Function or Mapping

```python
data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon',
                              'Pastrami', 'corned beef', 'Bacon',
                              'pastrami', 'honey ham', 'nova lox'],
                     'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
# Create mapping dict
meat_to_animal = {
  'bacon': 'pig',
  'pulled pork': 'pig',
  'pastrami': 'cow',
  'corned beef': 'cow',
  'honey ham': 'pig',
  'nova lox': 'salmon'
}
```

```python
lowercased = data['food'].str.lower()
lowercased
```

```python
data['animal'] = lowercased.map(meat_to_animal)
data
```

### **Replacing Values**

```python
# in case there are some obvious wrong data values
data = pd.Series([1., -999., 2., -999., -1000., 3.])
data
```

```python
data.replace(-999, np.nan)
```

```python
# replace multiple values at once,
data.replace([-999, -1000], np.nan)
```

```python
# different replacement for each value
data.replace([-999, -1000], [np.nan, 0])
```

```python
# different replacement for each value
data.replace({-999: np.nan, -1000: 0})
```
